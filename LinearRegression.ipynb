```python
import pandas as pd
```

```python
df=pd.read_csv('ARP_MitM_dataset.csv')
df

```

```python
df2=df[['1.000000000000000000e+00','1.294000000000000000e+03','0.000000000000000000e+00','1.000000000000000000e+00.1','1.294000000000000000e+03.1','0.000000000000000000e+00.1','1.000000000000000000e+00.2','1.294000000000000000e+03.2','0.000000000000000000e+00.2','0.000000000000000000e+00.48']]
df2
```

```python
df_new=df2[:]
df_new
```

```python
#1) 0.000000000000000000e+00.48
#2) 0.000000000000000000e+00.2
df1=df_new.drop(['0.000000000000000000e+00.48'],axis=1)
y_values=df_new['0.000000000000000000e+00.48']
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
scaled_df = sc.fit_transform(df1)
scaled_df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test= train_test_split(scaled_df,y_values, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import confusion_matrix, accuracy_score
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train) 
```

```python
predicted= regressor.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```
